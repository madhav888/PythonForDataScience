# matplotlib

1. Scatter Plot (Relationship b/w two variables)
2. Histogram (Numerical Variables)
3. Bar Chart (Categorical Variables)

```python
import numpy as np
import pandas as pd
cd = pd.read_csv('Toyota.csv',index_col=0,na_values=["??","????"])
```

```python
cd.dropna(axis=0,inplace=True)
```

### Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
# scatter(x_axis, y_axis, color)

plt.scatter(cd['Age'], cd['Price'], c='red')

plt.title('Plot of Price vs Age')

# Labels
plt.xlabel('Age in Months')
plt.ylabel('Price in EUR')

plt.show()
```

### Histograms
Using Ranges/Bins for Numerical Variables

```python
plt.hist(cd['KM'])
plt.show()
```

```python
# Beautification

plt.hist(cd['KM'],
        color = 'green',
        edgecolor = 'white',
        bins = 5 )

plt.title('Hist of KM')

# Labels
plt.xlabel('KM')
plt.ylabel('Frequency')

plt.show()
```

### Bar Plot
for Categorical Data

```python
# Create Arrays beforehand and then pass to bar()

counts = [979,120,12]
fuelType = ('Petrol', 'Diesel', 'CNG')
index = np.arange(len(fuelType))

plt.bar(index, counts, color=['red', 'blue', 'cyan'])

plt.title('Bar Plot of Fuel Types')

# Labels
plt.xlabel('Fuel Types')
plt.ylabel('Frequency')

plt.show()
```

```python
# xTicks/X-Labels

counts = [979,120,12]
fuelType = ('Petrol', 'Diesel', 'CNG')
index = np.arange(len(fuelType))
index1 = [1,0,2]

plt.bar(index, counts, color=['red', 'blue', 'cyan'])

plt.title('Bar Plot of Fuel Types')

# Labels
plt.xlabel('Fuel Types')
plt.ylabel('Frequency')

#xTicks
plt.xticks(index, fuelType)

plt.show()
```

```python
# Change Order by shuffling index

counts = [979,120,12]
fuelType = ('Petrol', 'Diesel', 'CNG')
index = np.arange(len(fuelType))
index1 = [1,0,2] # Change Order by shuffling index

plt.bar(index1, counts, color=['red', 'blue', 'cyan'])

plt.title('Bar Plot of Fuel Types')

# Labels
plt.xlabel('Fuel Types')
plt.ylabel('Frequency')

#xTicks
plt.xticks(index1, fuelType)

plt.show()
```

# seaborn

1. Scatter Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cd = pd.read_csv('Toyota.csv',index_col=0,na_values=["??","????"])
# Remove missing values directly from "cd" dataframe
cd.dropna(axis=0,inplace=True)
```

```python
# Plot of Price vs Age with Default Arguments
sns.set(style= "darkgrid")
# With Default Regression Line
sns.regplot(x= cd['Age'], y= cd['Price'])
# X and Y Labels are already Picked Up from Dataframe
```

```python
# Plot of Price vs Age with Default Arguments
sns.set(style= "darkgrid")
# Without Regression Line
## Change Marker to change datapoints representation
sns.regplot(x= cd['Age'], y= cd['Price'], marker= '*',fit_reg= False)
```
